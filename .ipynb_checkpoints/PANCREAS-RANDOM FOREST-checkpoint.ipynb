```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d = pd.read_csv('pancreas.csv')
d
```

```python
d.replace('BPTB',1,inplace=True)
d.replace('ESP',2,inplace=True)
d.replace('LIV',3,inplace=True)
d.replace('UCL',4,inplace=True)

d.replace('M',1,inplace=True)#MALE=1
d.replace('F',0,inplace=True)#FEMALE=2
d
```

```python
d.isna().any()
```

```python
#IA, IB, II, IIA, IIB, III, IV-cancer types
d['stage'].replace(np.nan,0,inplace=True)
d.replace('I',1,inplace=True)
d.replace('IA',2,inplace=True)
d.replace('IB',3,inplace=True)
d.replace('II',4,inplace=True)
d.replace('IIA',5,inplace=True)
d.replace('IIB',6,inplace=True)
d.replace('III',7,inplace=True)
d.replace('IV',8,inplace=True)
```

```python
plasma_mean=d['plasma_CA19_9'].mean()
reg1_mean=d['REG1A'].mean()

d['plasma_CA19_9'].replace(np.nan,plasma_mean,inplace=True)
d['REG1A'].replace(np.nan,reg1_mean,inplace=True)
d
```

```python
d.isnull().any()
```

```python
d=d.drop(['sample_id'],axis=1)
d=d.drop(['sample_origin'],axis=1)
d=d.drop(['patient_cohort'],axis=1)
d=d.drop(['benign_sample_diagnosis'],axis=1)
d.to_csv('pancreas_changed.csv')
```

```python
X = d.iloc[:, :-1].values
y = d.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
